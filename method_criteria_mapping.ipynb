```python
criteria = criteria_stage.index.tolist()
contexts = criteria_context.columns.values.tolist()
methods = method_criteria.index.tolist()
stages = criteria_stage.columns.values.tolist()

#create an empty dict to save stage-method arrays per each SDG
dict_stage_method = {}

#Calculate stage-method array and save it in dict
for i,context in enumerate(contexts): 
    dict_stage_method[context] = np.empty([len(stages), len(methods)])
    for j in range(len(stages)):        #iterate between stages     
        for k in range(len(methods)):     #iterate between methods
            sum_stage_method = 0
            for m in range(len(criteria)):  #iterate between criteria
                sum_stage_method += 1-abs((criteria_context.iloc[m,i]-method_criteria.iloc[k,m])/2)*criteria_stage.iloc[m,j]
            dict_stage_method[context][j,k] = sum_stage_method/len(criteria)
    
```

```python
#create a shared cbar for heatmap
import matplotlib.pyplot as plt
import matplotlib.colors
import seaborn as sns

fig, axes = plt.subplots(1, 2, sharex=False, sharey=True, figsize=(7, 13))
sns.set(font_scale=1.2)


#create colourmap for heatmaps based on context colour
context_colour = {'Goulburn Murray case':'YlGnBu', 'Forrest/Otways case':'YlGnBu'}


#Draw heatmaps
for i, (context, ax) in enumerate(zip(contexts, axes.flat)):
    pd_stage_method = pd.DataFrame.from_records(dict_stage_method[context], columns=methods, index=stages)
    pd_stage_method_t = pd_stage_method.transpose()
    
    labels = pd.DataFrame(index=methods, columns=stages)
    labels = labels.replace(np.nan, '', regex=True)

    for i, stage in enumerate(stages):
        top5 = list(pd_stage_method_t.nlargest(5, stage).index)
        for j, method in enumerate(top5):
            labels.loc[method, stage] = '+'
            

    sns.heatmap(pd_stage_method_t, center=.8, vmin=.6, vmax=1, annot=labels, annot_kws={'color':'white'}, fmt = '', cmap=context_colour[context], 
                cbar_kws = dict(use_gridspec=False,location="bottom", pad=0.15, anchor=(0, -.25)), 
                linewidths=4, xticklabels=True, yticklabels=True, ax=ax)
    ax.tick_params(axis='x', rotation=90)

    
    ax.xaxis.tick_top()
    ax.tick_params(left=False, bottom=False, top=False)
    
plt.subplots_adjust(wspace=.2, hspace=.09)


plt.savefig('{}/fig{}.png'.format(r'C:\Users\moallemie\OneDrive - Deakin University\My Library\Deakin\Publications_Deakin\Method selection\figs', '_heatplot'), 
            dpi=300, bbox_inches='tight')
```
